## Bitbucket

Jupyter/IPython notebook rendering is not supported. Support has been requested in [issue #11318](https://bitbucket.org/site/master/issues/11318/enable-rich-rendering-of-jupyter-notebooks) on the public issue tracker.

## GitLab and GitLab CE

### Inline LaTeX display

The first inline math block is displayed at the end of the line rather than where it should be. The remaining math blocks are not rendered.

This is tracked by [issue #30900](https://gitlab.com/gitlab-org/gitlab-ce/issues/30900) at the GitLab CE repository.

### Multiple outputs

The first code cell above has five output parts: a `stdout` block, an image block, another `stdout` block, another image block, and finally an execution result block. On GitLab, only the execution result block is shown.

The second code cell has no execution result block, but only two image blocks. On GitLab, only the first image block is shown.

[Issue #31910](https://gitlab.com/gitlab-org/gitlab-ce/issues/31910) has been opened to track this problem.
